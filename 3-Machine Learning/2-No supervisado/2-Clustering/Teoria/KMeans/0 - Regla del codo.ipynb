# El m√©todo del codo y el factor de la silueta del clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.cm as cm
```

```python
x1 = np.array([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
X = np.array(list(zip(x1,x2))).reshape(len(x1), 2)
```

```python
X
```

```python
plt.plot()
plt.xlim([0,10])
plt.ylim([0,10])
plt.title("Dataset a clasificar")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x1,x2)
# plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='red')
plt.show()
```

```python
from sklearn.cluster import KMeans

for k in range(1, 10):
    print("\n\n")
    print(f"K={k}")
    km = KMeans(n_clusters=k, n_init=10, random_state=0)
    km.fit(X)

    print(km.cluster_centers_)
    print(km.labels_)
    print(km.inertia_)

    plt.plot()
    plt.xlim([0,10])
    plt.ylim([0,10])
    plt.title(f"K={k}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.scatter(x1,x2)
    plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='red')
    plt.show()
```

```python
max_k = 10## maximo n√∫mero de clusters que vamos a crear
K = range(1,max_k)
ssw = []

for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    ssw.append(kmeanModel.inertia_)
    
#Representaci√≥n del codo
plt.plot(K, ssw, "bx-")
plt.xlabel("k")
plt.ylabel("SSw(k)")
plt.title("La t√©cnica del codo para encontrar el k √≥ptimo")
plt.show()    
```
