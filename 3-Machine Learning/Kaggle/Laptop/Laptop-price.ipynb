```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/train.csv')
```

```python
df
```

```python
X = df[['Pulgadas', 'Tipo']]
y = df['Precio']
```

```python
X = pd.get_dummies(X, columns=['Tipo'], drop_first=True)
X
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import get_scorer_names

- np.mean(cross_val_score(lm, X, y, scoring='neg_root_mean_squared_error'))
```

```python
df_test = pd.read_csv('data/test.csv')
```

```python
df_test
```

```python
X_test = df_test[['Pulgadas', 'Tipo']]
X_test = pd.get_dummies(X_test, columns=['Tipo'], drop_first=True)
```

```python
X_test
```

```python
y_test = lm.predict(X_test)
```

```python
df_sub = pd.DataFrame({'Id': df_test.Id, 'Precio':y_test})
df_sub
```

```python
df_sub.to_csv('submissions/submission_1.csv', index=False)
```
