Primero, cogemos todo el conjunto que tengamos de train. Este conjunto lo dividiremos en:
1. Train (etiquetado) para el alumno
2. Test (sin etiquetar) parael alumno. Los labels irán a Kaggle donde se hará un split del todo el conjunto de test
    - 50% para el score privado
    - 50% para el score público

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/cs-training.csv').reset_index()
df.rename(columns = {'index': 'Id'}, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

# Split
## Train

70% Para train

```python
np.random.seed(17)
df['split'] = np.random.randn(df.shape[0], 1)

msk = np.random.rand(len(df)) <= 0.7
train = df[msk]
test = df[~msk]

train = train.drop(columns='split')
train.head()
```

```python
test.head()
```

## Test

```python
test2 = test.drop(columns=['SeriousDlqin2yrs', 'split'])
test2.head()
```

```python
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
# Problema desbalanceado. Veamos cómo queda
test['SeriousDlqin2yrs'].value_counts()
```

## Answer file
Labels del conjunto de test que va a ir a Kaggle. Todo el conjunto objenido de Test, con el Id del orden que llevan + el label

```python
answer_key = test['SeriousDlqin2yrs']
answer_key = pd.DataFrame(answer_key)
answer_key['Id'] = answer_key.index
answer_key = answer_key[['Id', 'SeriousDlqin2yrs']]
answer_key.head()
```

## Sample submission file
Hace falta un sample submission file en el conjunto de datos.

Formato:
- Id: Igual que el de test
- SeriousDlqin2yrs: como se va a usar AUC, se necesitan las probabilidades, por lo que en este caso será un numero random entre 0 y 1

```python
sample_submission = answer_key.copy()
sample_submission['SeriousDlqin2yrs'] = np.random.random(len(answer_key))
sample_submission.head()
```

## Save files

```python
train.to_csv('kaggle_data/train.csv', index=False)
test2.to_csv('kaggle_data/test.csv', index=False)
answer_key.to_csv('kaggle_data/answer_key.csv', index=False)
sample_submission.to_csv('kaggle_data/sample_submission.csv', index=False)
```
