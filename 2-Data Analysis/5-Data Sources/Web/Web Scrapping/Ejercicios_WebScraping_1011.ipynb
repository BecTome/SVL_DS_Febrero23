# Miniproject: Obtención de Cotizaciones de Forex

En la web https://www.myfxdata.io podemos extraer las cotizaciones diarias de Forex. Se pide crear una función que, dado el cambio deseado (ej: EUR/USD), año y mes devuelva la tabla de cotizaciones en un DataFrame.

```python
url = 'https://myfxdata.io/priceaction/daily.php?D1_Pares=EURUSD&D1_Anyos=2021&D1_Meses=3'

resp = requests.get(url)
soup = BeautifulSoup(resp.text)
```

```python
table = soup.find('table', attrs={'class':'sample'})
print(table.prettify())
```

```python
# Primero sacamos las etiquetas de fila de la tabla
ls_rows = table.find_all('tr')

# Dentro de cada fila, queremos sacar las etiquetas de columna
ls_contents = []
for tag in ls_rows:
    # Dentro de cada etiqueta de columna, queremos sacar su texto
    ls_subtag = []
    for subtag in tag.find_all('td'):
        ls_subtag.append(subtag.string)

    ls_contents.append(ls_subtag)

ls_contents
```

```python
import pandas as pd
df = pd.DataFrame(ls_contents[1:], columns=ls_contents[0])
df = df.dropna(how='all')
df.head()
```

```python
df.shape
```

```python
def get_fxdata_year(year):
    ls_urls = [f'https://myfxdata.io/priceaction/daily.php?D1_Pares=EURUSD&D1_Anyos={year}&D1_Meses={month}' for month in range(1, 13)]
    ls_df = []
    for url in ls_urls:
        resp = requests.get(url)
        soup = BeautifulSoup(resp.text)
        table = soup.find('table', attrs={'class':'sample'})
        
        # Primero sacamos las etiquetas de fila de la tabla
        ls_rows = table.find_all('tr')

        if len(ls_rows)>4:
            # Dentro de cada fila, queremos sacar las etiquetas de columna
            ls_contents = []
            for tag in ls_rows:
                # Dentro de cada etiqueta de columna, queremos sacar su texto
                ls_subtag = []
                for subtag in tag.find_all('td'):
                    ls_subtag.append(subtag.string)

                ls_contents.append(ls_subtag)

            df = pd.DataFrame(ls_contents[1:], columns=ls_contents[0])
            df = df.dropna(how='all')

            ls_df.append(df)

        df_tot = pd.concat(ls_df, ignore_index=True)
    return df_tot
```
