

# Plotly Chart Studio
## Ejercicio Tips
En este ejercicio vas a desarrollar un dashboard interactivo en Plotly sin escribir una línea de código. Para realizar el ejercicio te apoyarás en la herramienta Chart Studio de Plotly. Tendrás que registrarte en su página web y realizar todo el desarrollo desde la misma.

Utiliza el dataset de propinas, cuyo enlace te dejo abajo. Se trata de responder a preguntas como:
* ¿Qué relación tienen lo que se paga de propina vs lo que cuesta comer?
* ¿Cuándo se paga más propia? ¿Para comer o para cenar?
* ¿Qué días de la semana se paga más propina? ¿Pagan más los hombres o las mujeres?

En el apartado de enlaces tienes también un ejemplo de cómo sería un dashboard de este estilo con el mismo dataset. Apoyate en ese dashboard para crear tus primeras gráficas y después continúa el análisis exploratorio con otras nuevas.
