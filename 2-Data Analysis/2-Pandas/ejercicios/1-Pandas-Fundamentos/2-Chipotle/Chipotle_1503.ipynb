### Paso 1. Importar las bibliotecas necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Paso 2. Importa el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) y asígnalos a una variable llamada chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
```

```python
chipo
```

### Paso 3. Nombre del producto de valor máximo

```python
chipo2 = chipo.copy()
chipo2["item_price"] = chipo2["item_price"].str.replace("$", "", regex=False)
chipo2["item_price"] = chipo2["item_price"].str.replace(" ", "", regex=False)
chipo2["item_price"] = chipo2["item_price"].astype(float)
```

```python
chipo["item_price"].dtype
```

```python
chipo2["item_price"].dtype
```

```python
chipo2.loc[chipo2["item_price"] == chipo2["item_price"].max(), : ]
```

### Paso 4. ¿Cuántos productos cuestan más de 10 dólares?

```python
chipo2.loc[chipo2["item_price"] > 10, ["item_name"]].count()
```

Extra: Cuántos pedidos se han hecho con un producto de más de 10$

```python
chipo2.loc[chipo2["item_price"] > 10, "order_id"].nunique()
```

### Paso 5. ¿Cuál es el precio de cada artículo? Obtenga sólo item_name y unit_price

```python
chipo2["quantity"]>=1
```

```python
chipo2.loc[(chipo2["item_name"]=="Canned Soft Drink")&(chipo2["choice_description"] == "[Coke]")]#Canned Soft Drink), :]#.sort_values(by="item_name")
```

```python
chipo2["unit_price"] = chipo2["item_price"] / chipo2["quantity"]
```

```python
chipo2.loc[(chipo2["item_name"]=="Canned Soft Drink")&(chipo2["choice_description"] == "[Coke]")]#Canned Soft Drink), :]#.sort_values(by="item_name")
```

```python
chipo2["unit_price"] = chipo2["item_price"] / chipo2["quantity"]
menu = chipo2.loc[:, ["item_name", "choice_description", "unit_price"]].drop_duplicates()
menu
```

### Paso 6. Ordenar por el nombre del artículo

```python
menu.sort_values(by=["item_name", "choice_description"])
```

### Paso 7. ¿Cuál fue la cantidad del artículo más caro pedido? 2 formas

```python
menu.sort_values(by="unit_price", ascending=False)
```

```python
menu[menu["unit_price"] == menu["unit_price"].max()]
```

```python
chipo2.loc[:, ["item_name", "choice_description", "quantity", "item_price", "unit_price"]]\
            .sort_values(by=["unit_price", "quantity"], ascending=[False, False])
```

```python
dff = chipo2.groupby(["item_name", "choice_description"])\
            .aggregate({"quantity": "sum", "unit_price":"max"})\
            .reset_index()

dff.sort_values(by="quantity", ascending=False).head(10)
```

## Paso 7.1: Cantidad de ARTICULOS del pedido MAS CARO

```python
df_orders = chipo2.groupby("order_id")[["item_price", "quantity"]].sum().reset_index()
df_orders.sort_values(by="item_price", ascending=False)
```

```python
def csoda(x):
    return x.loc[x["item_name"] == "Canned Soda", ["quantity"]].sum() > 1
chipo2.groupby("order_id").filter(csoda)
```
