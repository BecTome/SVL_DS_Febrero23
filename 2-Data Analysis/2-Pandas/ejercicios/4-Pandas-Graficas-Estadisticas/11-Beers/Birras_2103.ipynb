## Ejercicio 11
No queda muy claro cuánto de dispar, de dispersa, es la medida de IBU. Mediante estadística descriptiva (una o dos métricas, sin gráficas), razona la dispersión de la variable IBU.

```python
std_beers = df_beers["ibu"].std()
mean_beers = df_beers["ibu"].mean()

print(f"La media es {mean_beers:.2f} y la desviacion estandar es {std_beers:.2f}")
print(f"El amargor de las cervezas se encuentra entre {(mean_beers - std_beers):.2f}\
y {(mean_beers + std_beers):.2f}")
```

```python
q1 = df_beers["ibu"].quantile(.25)
q3 = df_beers["ibu"].quantile(.75)

iqr = q3 - q1
print(f"El amargor de las cervezas se encuentra entre {q1:.2f} y {q3:.2f}")
```
