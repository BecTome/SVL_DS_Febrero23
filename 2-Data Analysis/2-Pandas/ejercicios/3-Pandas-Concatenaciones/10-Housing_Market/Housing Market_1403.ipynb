# Mercado inmobiliario

### Introducción:

En esta oportunidad crearemos nuestro propio conjunto de datos con números ficticios para describir un mercado inmobiliario. Como vamos a crear datos aleatorios, no intentes razonar sobre los números.

### Paso 1. Importa las librerías necesarias

```python
import pandas as pd
import numpy as np
```

### Paso 2. Crea 3 Series diferentes, cada una con una longitud de 100, de la siguiente manera:
1. La primera un número aleatorio del 1 al 4
2. La segunda un número aleatorio del 1 al 3
3. La tercera un número aleatorio de 10,000 a 30,000

```python
arr_1 = pd.Series(np.random.randint(1, 5, 100))
arr_2 = pd.Series(np.random.randint(1, 4, 100))
arr_3 = pd.Series(np.random.randint(1E4, 3E4 + 1, 100))
print(arr_1.shape, arr_2.shape, arr_3.shape)
```

### Paso 3. Creamos un DataFrame uniendo las Series por columna

```python
df = pd.concat([arr_1, arr_2, arr_3], axis=1)
df.head()
```

### Paso 4. Cambia el nombre de las columnas a bedrs, bathrs, price_sqr_meter

```python
new_names = ["bedrs", "bathrs", "price_sqr_meter"]
df.columns = new_names
df
```

```python
df = df.rename(columns={0: "bedrs", 1: "bathrs", 2:"price_sqr_meter"})
df
```

```python
new_names = ["bedrs", "bathrs", "price_sqr_meter"]
df = df.set_axis(new_names, axis=1)
df
```

### Paso 5. Creamos un DataFrame de columna única con los valores de las 3 Series y lo asignamos a 'bigcolumn'

```python
bigcolumn = pd.concat([arr_1, arr_2, arr_3])
bigcolumn
```
