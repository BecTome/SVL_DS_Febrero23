## Parentesis vs corchetes
1. **Corchetes** (Selección de filas o columnas):
    - Listas y tuplas: indexing y slicing
    - Numpy: index y slicing
    - Dataframe: inexing y slicing + loc/iloc
2. **Paréntesis**: para métodos y funciones.
    - Si viene de un objeto: es un metodo. objeto.metodo()
    - Si se ejecuta tal cual, e suna función o una clase.

## Group by
Cuando queremos calcular une stadistico para cada grupo de una variable categorica

````python
# Media de ventas
df['Sales'].mean()

# Media de ventas para cada ciudad
# df.groupby('City').mean()['Sales']

# # Media de ventas por segmento y ciudad
df.groupby(['City', 'Segment']).mean()['Sales'].unstack(level="Segment")#.reset_index(0) # 0 o 1
````

## Filtrar, máscaras

````python
# LAS MASCARAS ESTÁN PARA FILTRAR FILAS
# Mascara de todo el DF
df.isnull()

# Mascara por una columna
df['State'] == 'Kentucky'

# Multiples condiciones
mask = (df['State'] == 'Kentucky') & (df['Ship Mode'] == 'Second Class')

# Multiples valores para una columna
ciudades = ['Henderson', 'Florence']
df[df['City'].isin(ciudades)]

# Aplicar una máscara
df[mask]

# Filtrar nulos. No hay en esta columna
df[df['City'].isnull()]
````

## Apply
Cuando hay que modificar valores de una columna

````python
# Una columna de entrada, una de salida
def mifun(x):
    return x*100000

df['Sales'].apply(mifun)
````

````python
df[['City', 'State']].apply(lambda x: x["City"] + "-" + x["State"], axis=1)
````

````python
# Varias columas de entrada, una salida
def mifun(x, y):
    return x + ' - ' + y

df.apply(lambda x: mifun(x['City'], x['State']), axis=1)
````

````python
# Una columa de entrada, varias de salida
def mifun(x):
    return pd.Series((x*10, x*1000))

df[['Sales_10', 'Sales_1000']] = df['Sales'].apply(mifun)
df.head()
````

````python
# Varias columnas de entrada, varias de salida
def mifun(x):
    return pd.Series((x*10, x*1000))

df[['Sales_10', 'Sales_1000']] = zip(*df['Sales', 'Profit'].apply(mifun, axis=1))
df.head()
````

## Concat, append, merge
Siempre hay una clave de cruce y unas columnas comunes. Se puede razonas los joins con teoria de conjuntos (bolas y nombres de clientes...)
1. **Concat** (mantiene los index originales):
    - Entre series, concatena los series linealmente
    - Entre DF: los oncatena verticalmente si coinciden en columnas
    - pd.concat([x, y], ignore_index=True) para que ignore el indice y ponga un RangeIndex nuevo.
2. **Append**: funciona como el concat. Existe por el nombre que tiene. Concatenado vertical.
3. **Merge**:
    - Realiza el join por las columnas comunes entre os dos dataframes
    - Por defecto es un inner join.
    - 'left_on' y 'right_on' para especificar las columnas de cruce
    - 'left_index' y  'right_index' por si se quiere hacer el cruce por indice
    - Se pueden mezclar en el cruce indice con columna
    - Outer es en realidad full outer join
4. **Join**:
    - Por defecto, lo hace por indice

```Python
df3 = pd.merge(df1, df2)
pd.merge(df1, df3, left_on="employee", right_on="name")
pd.merge(df1a, df2a, left_index=True, right_index=True)
```
