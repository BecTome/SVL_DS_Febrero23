```python
import numpy as np
import pandas as pd
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

## 4. Genere una consulta que cree un campo “co_ranking” que clasifique en orden ascendente las campañas según el total de venta, siendo 1 la campaña que más ha vendido y N la que menos. (N = Total de campañas)

```python
camp_dpto = pd.merge(CAMPANIA, DEPTO_CAMPANIA, how='left')
camp_art = pd.merge(camp_dpto, ARTICULO, how='left')
camp_tipo = pd.merge(camp_art, DEPARTAMENTO)
camp_venta = pd.merge(camp_tipo, VENTA, how='inner')
camp_venta
```

```python
camp_sum = camp_venta.groupby('ID_CAMPANIA')[['PRECIO']].sum()
sorted_camp_sum = camp_sum.sort_values(by='PRECIO', ascending=False)
sorted_camp_sum['co_ranking'] = range(1, len(sorted_camp_sum)+1)
sorted_camp_sum
```

## 10. Genere una o varias consultas que devuelvan la variación del total de venta entre años. Tomando como variación la siguiente fórmula $$\dfrac{Year_{n} – Year_{n-1}}{Year_{n-1}}$$ (Se valorará el control de valores 0 en la división o de indeterminaciones)

![tablas5](img/Tablas-5.png)

```python
venta_year = VENTA.copy()
venta_year.index = venta_year['FECHA_VENTA']
venta_year = venta_year.groupby(venta_year.index.year)[['PRECIO']].sum().reset_index()
venta_year
```

```python
venta_year['ANIO_1'] = venta_year['FECHA_VENTA'].shift()
venta_year
```

```python
venta_year['PRECIO_ANT'] = venta_year['PRECIO'].shift()
venta_year['VARIACION'] = (venta_year['PRECIO'] - venta_year['PRECIO_ANT']) / venta_year['PRECIO_ANT']
venta_year
```

```python
venta_year.rename(columns={'FECHA_VENTA':'ANIO_2'}, inplace=True)
```

```python
venta_year = venta_year[['ANIO_1', 'ANIO_2', 'VARIACION']].dropna()
venta_year['ANIO_1'] = venta_year['ANIO_1'].astype('int')
venta_year
```

```python
venta_year['VARIACION_PERCENT'] = venta_year['VARIACION'] * 10000 // 100
venta_year['VARIACION_PERCENT'] = venta_year['VARIACION_PERCENT'].astype(str) + "%"

venta_year
```
